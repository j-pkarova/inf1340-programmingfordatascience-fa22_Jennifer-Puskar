# Start Cleaning

```python
#Looks lik lots of NaN values for title/citations, with data header fields beginning around row 14. 
#Let's see the tail. 
UNmst_df1.tail(5)
```

```python
#Dropping rows to get to data. I have decided to drop the header columns for data cleaning (will rename myself).
UNmst_df1.drop(UNmst_df1.index[0:15], inplace=True)
UNmst_df1.head(5)
```

```python
#Reset index 
UNmst_df1.reset_index(inplace=True,drop=True)
UNmst_df1.head(5)
```

```python
#Let's look at the shape of the data now. It has less rows now (266 vs. 280)
UNmst_df1.shape
```

```python
#We need headers.
UNmst_df1.columns = ['Sort order','Major area, region, country','Notes','Country code','Type of data','T1990','T1995','T2000','T2005','T2010','T2015','M1990','M1995','M2000','M2005','M2010','M2015','F1990','F1995','F2000','F2005','F2010','F2015'
            ]
UNmst_df1.columns
```

```python
UNmst_df1.info
```

```python
#Let's transpose columns (years).
UNmst_df1 = UNmst_df1.melt(id_vars = ['Sort order','Major area, region, country','Notes','Country code','Type of data'],
              var_name = ["SexYear"],
              value_name = "Migrant Population",)
UNmst_df1.head(5)
```

```python
#Sex and year need to be separated.
UNmst_df1=(UNmst_df1.assign(Sex = lambda x: x.SexYear.str[0].astype(str), Year = lambda x: x.SexYear.str[1:].astype(str)).drop("SexYear",axis=1))
UNmst_df1.head(5)
```

```python
#We need to distinguish between countries and regions. This could help with Pivots of table splits later.
UNmst_df1['Major Area/Region vs. Country?'] = np.where(UNmst_df1['Type of data'].notna(), 'Country', 'Major Area or Region')
UNmst_df1.head(15)
```

```python
#Used if else to make seperate columns for each type of data. 
#There are more efficient ways to do this, but I wasn't able to get lambda to work. Practicing if-else:
resultB=[]
for value in UNmst_df1["Type of data"]:
  if value == 'B'or value == 'C B' or value == 'B R' or value == 'B I' or value == 'C B R':
    resultB.append("B")
  else:
    resultB.append(" ") #What should I put here?
UNmst_df1["B Data (Foreign-born)"]=resultB
```

```python
resultC=[]
for value in UNmst_df1["Type of data"]:
  if value == 'C'or value == 'C B' or value == 'C B R' or value == 'C R':
    resultC.append("C")
  else:
    resultC.append(" ")
UNmst_df1["C Data (Foreign citizens)"]=resultC
```

```python
resultR=[]
for value in UNmst_df1["Type of data"]:
  if value == 'B R'or value == 'C B R' or value == 'C R' or value == 'I R':
    resultR.append("R")
  else:
    resultR.append(" ")
UNmst_df1["R Data (Refugees)"]=resultR
```

```python
resultI=[]
for value in UNmst_df1["Type of data"]:
  if value == 'I'or value == 'I R':
    resultI.append("I")
  else:
    resultI.append(" ")
UNmst_df1["I Data (Imputation)"]=resultI
```
